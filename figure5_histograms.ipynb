```python
power2_antilog = []
power3_antilog = []
power2_mean = []
power3_mean = []

for file_number in file_range:

    file2 = files2fce[file_number]
    df2 = pd.read_csv(files2fce[file_number],header=None,delim_whitespace=True)
    power2 = df2[2].to_numpy()
    
    for power in power2:
        power2_antilog.append(10**(power/100))
    
    power2_mean.append(np.mean(power2_antilog))
    
    file3 = files3fce[file_number]
    df3 = pd.read_csv(files3fce[file_number],header=None,delim_whitespace=True)
    power3 = df3[2].to_numpy()
    
    for power in power3:
        power3_antilog.append(10**(power/100))
    
    power3_mean.append(np.mean(power3_antilog))
```

```python
relative_power = 10*np.log10(np.array(power2_mean)/np.array(power3_mean))

relative_mean = np.mean(relative_power)
relative_std = np.std(relative_power)

fig = plt.figure(figsize = (8,8))
frequencies, bin_edges,pathches = plt.hist(relative_power,density=False,rwidth=0.8,color='purple',label='',bins=10,alpha=0.8)
#plt.plot(x,normal.pdf(x),lw=5,ls='--',color='green')
plt.ylabel('Counts',fontsize=30)
plt.yticks(fontsize=30)
plt.xlabel('10*log[P(2fce)/P(3fce)] (dB)',fontsize=30)
plt.xticks(fontsize=30,rotation=45)
plt.title('Relative Power Distribution',fontsize=30)
#plt.legend(prop={'size': 30})
plt.tight_layout()
#plt.savefig('power-distribution-relative-new.png')
```
