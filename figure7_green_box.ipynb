```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections
from collections import Counter
```

```python
def data_frame_3(path,files):
    
    output_folder = sorted(glob.glob("{}{}/*.txt".format(path,files)))
    df_all_rows = pd.concat([pd.read_csv(item,header=None,delim_whitespace=True) for item in output_folder],axis = 0,ignore_index=True)
    df = df_all_rows[df_all_rows[3] >= 751.]    
    input_azimuth_list = []
    input_elevation_list = []
    
    for full_path in output_folder:
        input_azimuth_list.append(int(full_path[33:36]))
        input_elevation_list.append(int(full_path[39:41]))
    
    input_azimuth = np.array(input_azimuth_list)
    input_elevation = np.array(input_elevation_list)
    power = df[3].to_numpy(dtype=float)
    output_azimuth_180 = df[4].to_numpy(dtype=float)
    output_azimuth_list = [angle + 360 if angle < 0 else angle for angle in output_azimuth_180]
    output_azimuth = np.array(output_azimuth_list)
    output_elevation = df[5].to_numpy(dtype=float)
    
    return(input_azimuth,input_elevation,power,output_azimuth,output_elevation)
```

```python
'Opening Aliased Data:'

path = 'aliased_dataset/'
files = '3_90_MHz/files'
input_azimuth,input_elevation,power,output_azimuth,output_elevation = data_frame_3(path,files)

X = np.reshape(input_azimuth,(181,41))
Y = np.reshape(input_elevation,(181,41))
Z_az_3 = np.reshape(output_azimuth,(181,41))

X = np.reshape(input_azimuth,(181,41))
Y = np.reshape(input_elevation,(181,41))
Z_el_3 = np.reshape(output_elevation,(181,41))
```

```python
'Opening Measured Data:'

filenumber = 0
filepath = 'figure7_dataset/green_box/'
filepath_len = len(filepath)
files2fce = glob.glob('{}*-2.csv'.format(filepath))
files3fce = glob.glob('{}*-3.csv'.format(filepath))
file2 = files2fce[filenumber]
file3 = files3fce[filenumber]
filename = file3[ filepath_len : filepath_len + 18 ]
```

```python
'Data Framing:'

df2 = pd.read_csv(file2,header=None,delim_whitespace=True)
df3 = pd.read_csv(file3,header=None,delim_whitespace=True)
freq2 = df2[1].to_numpy()
freq3 = df3[1].to_numpy()
ele2 = df2[3].to_numpy()
azi2 = df2[4].to_numpy()
ele3 = df3[3].to_numpy()
azi3 = df3[4].to_numpy()
data_degrees2 = list(zip(azi2,ele2))
data_degrees3 = list(zip(azi3,ele3))
data_radians2 = [(np.deg2rad(a), r) for a, r in data_degrees2]
data_radians3 = [(np.deg2rad(a), r) for a, r in data_degrees3]
data_array2 = np.array(data_radians2)
data_array3 = np.array(data_radians3)
```

```python
def find_alias_3(test_el,test_az,text_output=False):
    
#create testpoint
    testpoint = np.array([test_el,test_az])
# initialize filtered arrays to all zeros
    elfiltered = np.zeros((181,41),dtype=int)
    azfiltered = np.zeros((181,41),dtype=int)
# filtering process
    for i in range(181):
        for j in range(41):
            if abs(Z_el_3[i,j]-testpoint[0]) <= 1:
                elfiltered[i,j] = 1
            if abs(Z_az_3[i,j] - testpoint[1]) <= 1:
                azfiltered[i,j] = 1
# finding solutions
    solution = []
    count = 0 #counter for the number of solutions
    for i in range(181):
        for j in range(41):
            if elfiltered[i,j] == 1 and azfiltered[i,j]==1:
                count += 1
                solution.append([5+j*2,2*i])
    if text_output == True:
        # output solutions
        if count == 0:
            print("There are no solutions.")
        else:
            print(f"Total solutions found: {count}") 
            for sol in solution:
                print(f"Solution: Elevation {sol[0]}, Azimuth {sol[1]}")
    return(testpoint, count, solution)
```

```python
'Saving All Solutions as a List of Dictionaries:'

all_solutions = []
for (el, az) in zip(ele3,azi3):
    testpoint, count, solution = find_alias_3(el,az)
    all_solutions.append({"real_el_and_az":testpoint,"count":count, "solutions":solution})
```

```python
test = []
for entry in all_solutions:
    for solution in entry["solutions"]:
        test.append(solution)
el_solns_3 = [item[0] for item in test]
az_solns = [item[1] for item in test]

az_solns_radians_3 = np.deg2rad(az_solns)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='polar')
ax.set_rlim(bottom=90, top=0)
ax.xaxis.grid(True,color='dimgray')
ax.yaxis.grid(True,color='dimgray')
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
ax.set_yticklabels(['0','','','30','','','60','','','90'],fontsize=30,color='silver')
ax.set_rlabel_position(45.0)
ax.set_xticklabels(['N', '', 'E', '', '', '', '', ''],fontsize=30,color='k')

ax.scatter(data_array2[:, 0], data_array2[:, 1], color='magenta', s=100, alpha=1, label='2.80 MHz')
ax.scatter(az_solns_radians_3, el_solns_3, color='b', s=100, alpha=1, label='3.90 MHz Aliased')
ax.scatter(data_array3[:, 0], data_array3[:, 1], color='cyan', s=100, alpha=1, label='3.90 MHz')

ax.legend(bbox_to_anchor=(0.86, 1), loc='upper left', borderaxespad=0)
#plt.savefig('figures/figure7/figure7_polar_plot_green_box.png',transparent=True)
plt.show()
```

```python
#Cyan: want to grey out 0:180
az_cyan2grey = []
el_cyan2grey = []
az_cyan = []
el_cyan = []

for i,angle in enumerate(azi3):
    if (angle>0 and angle<90) or (angle>270 and angle<360):
        az_cyan2grey.append(angle)
        el_cyan2grey.append(ele3[i])
    else:
        az_cyan.append(angle)
        el_cyan.append(ele3[i])
az_cyan2grey_rad = np.deg2rad(az_cyan2grey)
az_cyan_rad = np.deg2rad(az_cyan)


#Blue: want to grey out 0:90
az_solns_deg_3 = np.rad2deg(az_solns_radians_3)
az_blue2grey = []
el_blue2grey = []
az_blue = []
el_blue = []

for i,angle in enumerate(az_solns_deg_3):
    if (angle>0 and angle<90) or (angle>270 and angle<360):
        az_blue2grey.append(angle)
        el_blue2grey.append(el_solns_3[i])
    else:
        az_blue.append(angle)
        el_blue.append(el_solns_3[i])
az_blue2grey_rad = np.deg2rad(az_blue2grey)
az_blue_rad = np.deg2rad(az_blue)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='polar')
ax.set_rlim(bottom=90, top=0)
ax.xaxis.grid(True,color='dimgray')
ax.yaxis.grid(True,color='dimgray')
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)
ax.set_yticklabels(['0','','','30','','','60','','','90'],fontsize=30,color='silver')
ax.set_rlabel_position(45.0)
ax.set_xticklabels(['N', '', 'E', '', '', '', '', ''],fontsize=30,color='k')

ax.scatter(data_array2[:, 0], data_array2[:, 1], color='magenta', s=100, alpha=1, label='2.80 MHz')
ax.scatter(az_blue2grey_rad, el_blue2grey, color='grey', s=100, alpha=1)
ax.scatter(az_blue_rad, el_blue, color='blue', s=100, alpha=1)
ax.scatter(az_cyan2grey_rad, el_cyan2grey, color='grey', s=100, alpha=1)
ax.scatter(az_cyan_rad, el_cyan, color='cyan', s=100, alpha=1)

#plt.savefig('figures/figure7/figure7_polar_plot_green_box_grey.png',transparent=True)
plt.show()
```
